```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
data = pd.read_csv("overledenen_cbs_data.csv") 
data.head() # prints first 5 rows
```

```python
title = 'Overleden personen per week van 2011 tot nu, bron CBS (17-10-2020)'
xlabel = 'Week 2-51' # first and last week of the year can be shorter than 7 days
ylabel = 'Overleden personen'

plt.figure(num=None, figsize=(20, 10))
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)

for year in [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]:
    
    weeks = []
    total = []
    age_0_65 = []
    age_65_80 = []
    age_80up = []
    
    for row in data.iterrows():        
        if row[1]['Jaar'] == year:
            weeks.append(row[1]['Week'])
            total.append(row[1]['Totaal'])
            age_0_65.append(row[1]['0 tot 65 jaar'])
            age_65_80.append(row[1]['65 tot 80 jaar'])
            age_80up.append(row[1]['80 jaar of ouder'])
            
    plt.plot(weeks, total, label=year)

plt.xlim(2, 51) # first and last week of the year can be shorter than 7 days
plt.legend()
plt.show()
```

```python
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2,2, figsize=(20, 10))

plt.suptitle(title, fontsize=16)
ax1.set_title('Totaal (alle leeftijdsgroepen)', fontweight="bold")
ax2.set_title('80 jaar of ouder', fontweight="bold")
ax3.set_title('65 tot 80 jaar', fontweight="bold")
ax4.set_title('0 tot 65 jaar', fontweight="bold")

for year in [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]:
    
    weeks = []
    total = []
    age_0_65 = []
    age_65_80 = []
    age_80up = []
    
    for row in data.iterrows():        
        if row[1]['Jaar'] == year:
            weeks.append(row[1]['Week'])
            total.append(row[1]['Totaal'])
            age_0_65.append(row[1]['0 tot 65 jaar'])
            age_65_80.append(row[1]['65 tot 80 jaar'])
            age_80up.append(row[1]['80 jaar of ouder'])
            
    ax1.plot(weeks, total, label=year)
    ax2.plot(weeks, age_80up, label=year)
    ax3.plot(weeks, age_65_80, label=year)
    ax4.plot(weeks, age_0_65, label=year) 

for ax in [ax1, ax2, ax3, ax4]:
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_xlim([2, 51]) # first and last week of the year can be shorter than 7 days
    ax.set_ylim([300, 5200])
    ax.legend()

plt.show()
```
